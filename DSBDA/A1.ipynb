```python
import pandas as pd
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [23,34,45],
        'Gender': ['F', 'M', 'M'],
        'Marks': [89,98,'NaN']
    }

```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
n = avg = 0
for mark in df['Marks']:
  if mark is not 'NaN':
    avg += mark
    n += 1
avg /= n
print("The average marks are: ",avg)
```

```python
df.replace('NaN', value = avg)
```

```python
df['Gender'] = df['Gender'].map({'M' : 0, 'F' : 1}).astype(int)
```
